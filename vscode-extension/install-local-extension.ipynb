# Install the Ronin VS Code extension from local source

This notebook documents the steps needed to install the local VS Code extension package from the `vscode-extension` folder.

## 1. Open Extension Folder in VS Code

Open the local extension project folder in VS Code to prepare for installation.

- In VS Code, use `File` → `Open Folder...` and select `c:\\cygwin64\\home\\BTCES\\roninaiagent\\vscode-extension`.
- Confirm the extension files and `package.json` are visible in the Explorer.

## 2. Install Dependencies

Install the extension dependencies before building.

Run the package manager command from the extension folder:

```bash
cd "c:\\cygwin64\\home\\BTCES\\roninaiagent\\vscode-extension"
npm install
```

If you use `yarn` instead, replace with `yarn install`.

## 3. Build the Extension

Build the extension using the configured npm script.

```bash
cd "c:\\cygwin64\\home\\BTCES\\roninaiagent\\vscode-extension"
npm run compile
```

If the project defines `npm run build` or `npm run watch`, use that instead.

## 4. Package the Extension as VSIX

Create a `.vsix` file from the extension source.

```bash
cd "c:\\cygwin64\\home\\BTCES\\roninaiagent\\vscode-extension"
npx @vscode/vsce package --out ronin-vscode-0.1.5.vsix
```

If the extension package script is available, you can also run:

```bash
npm run package
```


## 5. Install the VSIX Locally

Install the generated `.vsix` file into VS Code using the CLI or the UI.

```bash
cd "c:\\cygwin64\\home\\BTCES\\roninaiagent\\vscode-extension"
code --install-extension ronin-vscode-0.1.5.vsix
```

If `code` is not on your PATH, use the VS Code command palette and choose `Extensions: Install from VSIX...`.

## 6. Launch Extension Development Host

Open the extension in the VS Code Extension Development Host to verify it runs locally.

Use the `Run and Debug` view and select `Launch Extension` or press `F5`.

Confirm that the Ronin Agent view appears and that login/chat actions work.

````python
import subprocess
import os

extension_dir = r'c:\cygwin64\home\BTCES\roninaiagent\vscode-extension'
command = ['cmd.exe', '/c', 'cd /d', extension_dir, '&&', 'npx', '@vscode/vsce', 'publish']
print('Publishing extension from:', extension_dir)
result = subprocess.run(command, capture_output=True, text=True)
print('returncode:', result.returncode)
print('stdout:')
print(result.stdout)
print('stderr:')
print(result.stderr)
````
